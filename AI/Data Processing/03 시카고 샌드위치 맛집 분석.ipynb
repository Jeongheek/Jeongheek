```python
# 메인페이지 데이터 수집 => 리스트에 저장
url_base='https://www.chicagomag.com'
url_sub='/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago'

# main 화면 수집
url=url_base+url_sub

import requests

request_headers={
    'User-Agent':'Mozilla/5.0'
}
resp=requests.get(url=url, headers=request_headers).text

# print(type(resp))
soup=BeautifulSoup(resp,'html.parser')

# 여기까지 메인화면 추출

ranks=[]
main_menus=[]
cafe_names=[]
url_adds=[]

list_soup=soup.find_all('div',class_='sammy') # 50개 들어있는 리스트

for item in list_soup:
    ranks.append(item.find(class_='sammyRank').text) # 순위만 추출
    tmp_str=item.find(class_='sammyListing').text # 메인메뉴\n카페명\n
    main_menus.append(tmp_str.split('\n')[0])
    cafe_names.append(tmp_str.split('\n')[1])
    
    tmp_url=item.find('a')['href']
    if tmp_url.startswith('https://'):
        url_adds.append(tmp_url)
    else:
        url_adds.append(url_base + tmp_url)
```

```python
import numpy as np
import pandas as pd

data=pd.DataFrame({
      'Rank':ranks
    , 'Menu':main_menus
    , 'Cafe':cafe_names
    , 'Url':url_adds
})
data.head(10)
```

```python
df = pd.read_csv(
    './시카고샌드위치데이터.csv'
    , encoding='utf-8'
    , index_col=0 # 첫번째 컬럼 인덱스로 보내라
)
df.head()
```

```python
url_str = df['Url'][0] # array object return
url_str
```

```python
# 위 코드까지 메인화면에 있는 데이터 추출
# 아래 코드에서 상세하면(50개)에 있는 가격, 주소 추출
# 1개 추출 : 1등 BLT 사이트의 가격, 주소 추출
import requests
from bs4 import BeautifulSoup

url_str=df['Url'][0] # 0 ~ 49

request_headers={
    'User-Agent':'Mozilla/5.0'
}
resp=requests.get(url=url_str, headers=request_headers).text # BLT 상세에서 전체 HTML resp

print(resp)
```

```python
# text => DOM 파싱
soup_tmp = BeautifulSoup(resp, 'html.parser') 

# 가격, 주소가 있는 영역 추출 <p class='addy'>가격, 주소 연락처, URL</p>
price_tmp=soup_tmp.find('p', class_='addy').get_text()
print(price_tmp) # Tag object => string 추출 => split
```

```python
# 위코드를 50번 반복
# tqdm 설치 : pip install tqdm
from tqdm import tqdm

# 가격, 주소 저장
price=[]
address=[]

for idx in tqdm(df.index):
    url_str=df['Url'][idx] # 0 ~ 49

    request_headers={
        'User-Agent':'Mozilla/5.0'
    }
    resp=requests.get(url=url_str, headers=request_headers).text # BLT 상세에서 전체 HTML resp
    
    soup=BeautifulSoup(resp, 'html.parser')
    
    get_str=soup.find('p', class_='addy').get_text()
    
    price.append(get_str.split()[0][:-1]) # '$10'
    address.append(' '.join(get_str.split()[1:-2])) # address 추출
```

```python
price, len(price)
```

```python
address, len(address)
```

```python
df.head()
```

```python
df['Price']=price
df['Address']=address

df.head()
```

```python
# df.loc[] : named index
# df.loc[행,열]

df = df.loc[:,['Rank','Cafe','Menu','Price','Address']]
df.head()
```

```python
# Rank 컬럼 => index로 보내기
df.set_index('Rank', inplace=True)
df.head()
```

```python
df.to_csv(
    './data/03. best_sandwiches_list_chicago2.csv'
    , encoding='utf-8'
)
```

```python
# lat, lng 구한다
# 주소에 Multiple => 주소가 없다

lat=[]
lng=[]

for idx in tqdm(df.index): # 50번 반복
    if df['Address'][idx] != 'Multiple':
        # 주소가 있다
        target_name=df.loc[idx,'Address'] + ' Chicago' # 시카고가 포함된 주소 
        gmaps_output=gmaps.geocode(target_name)
        location_output=gmaps_output[0].get('geometry')
        lat.append((location_output['location']['lat']))
        lng.append((location_output['location']['lng']))
    else:
        # 주소가 없다
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng)
```

```python
df['Lat']=lat
df['Lng']=lng
df.head()
```

```python
map=folium.Map(
    location=[df['Lat'].mean(), df['Lng'].mean()]
    , zoom_start=11
)

for idx in df.index:
    if df['Address'][idx] != 'Multiple':
       folium.Marker(
           [df['Lat'][idx],df['Lng'][idx]]
           , popup=df['Cafe'][idx]
       ).add_to(map)
    
map
```
